```python
especie_comun = arboles.nombre_com.value_counts().index[0]
print(f"La especie de arbol mas comun en CABA es: {especie_comun}")
```

```python
oeste = arboles.loc[arboles["long"].idxmin()]
print(f"El arbol mas al oeste de la ciudad es: {oeste['nombre_com']}, en {oeste['espacio_ve']}")
este = arboles.loc[arboles["long"].idxmax()]
print(f"El arbol mas al este de la ciudad es: {este['nombre_com']}, en {este['espacio_ve']}")
norte = arboles.loc[arboles["lat"].idxmax()]
print(f"El arbol mas al norte de la ciudad es: {norte['nombre_com']}, en {norte['espacio_ve']}")
sur = arboles.loc[arboles["lat"].idxmin()]
print(f"El arbol mas al sur de la ciudad es: {sur['nombre_com']}, en {sur['espacio_ve']}")
```

Mas al norte es mayor latitud mientras al sur, es menor latitud. \
Mas al este es mayor longitud mientras al oeste, es menor longitud.

```python
arbol_alto = arboles.loc[arboles['altura_tot'].idxmax()]
print("Arbol más alto de la CABA:")
print(f"Especie: {arbol_alto['nombre_com']}")
print(f"Nombre Científico: {arbol_alto['nombre_cie']}")
print(f"Espacio Verde: {arbol_alto['espacio_ve']}")
print(f"Altura: {arbol_alto['altura_tot']} mts")
print(f"Diámetro: {arbol_alto['diametro']} cm")
print(f"Inclinación: {arbol_alto['inclinacio']} Grados")
```

```python
norte_espacio_verde = arboles.loc[arboles["lat"].idxmax()]
print(f"El espacio verde más al norte de la ciudad es: {norte_espacio_verde['espacio_ve']}")
sur_espacio_verde = arboles.loc[arboles["lat"].idxmin()]
print(f"El espacio verde más al sur de la ciudad es: {sur_espacio_verde['espacio_ve']}")
```

f) ¿Cuáles son las especies identificadas como especies Nativas/autóctonas de la Ciudad de Buenos Aires? ¿Cuáles son las especies exóticas? \

```python
espacios_verdes_altos = arboles.groupby('espacio_ve')['altura_tot'].mean().sort_values(ascending=False)
espacio_mas_alto = espacios_verdes_altos.idxmax()
arboles_total = arboles[arboles['espacio_ve'] == espacio_mas_alto].shape[0]
especies_unicas = arboles[arboles['espacio_ve'] == espacio_mas_alto]['nombre_com'].unique()

print("El espacio verde que tiene los árboles más altos en promedio es: ", espacio_mas_alto, "con un promedio de:", round(espacios_verdes_altos.max(), 2), "mts")
print("Tiene", arboles_total, "árboles y la especie es:", especies_unicas[0])
```



---



```python
especies_nativas = arboles[arboles['origen'] == 'Nativo/Autóctono']['nombre_com'].unique()
especies_exoticas = arboles[arboles['origen'] == 'Exótico']['nombre_com'].unique()
salto_pagina=0
print("Las Especies Nativas son:\n")
for i in range(0,len(especies_nativas)):
  if i == len(especies_nativas)-1:
    print(especies_nativas[i], end=".\n\n")
  elif i == len(especies_nativas)-2:
    print(especies_nativas[i], end=" y ")
  else:
    print(especies_nativas[i], end=", ")
    salto_pagina = salto_pagina+len(especies_nativas[i])
    if salto_pagina+len(especies_nativas[i+1])>150:
      print("\n")
      salto_pagina=0
salto_pagina=0
print("\nLas Especies Exóticas son:\n")
for b in range(0,len(especies_exoticas)):
  if b == len(especies_exoticas)-1:
    print(especies_exoticas[b], end=".\n\n")
  elif  b == len(especies_exoticas)-2:
    print(especies_exoticas[b], end=" y ")
  else:
    print(especies_exoticas[b], end=", ")
    salto_pagina = salto_pagina+len(especies_exoticas[b])
    if salto_pagina+len(especies_exoticas[b+1])>150:
      print("\n")
      salto_pagina=0
```

i) Realiza un gráfico que permita ver las diez especies con menor altura promedio. Identifica estas especies con su nombre común/genérico. ¿Qué atributo tendría que estar en el eje y del gráfico? \

En ambos casos en necesario ordenar los datos.(Pista para ordenar podes usar np.sort())



```python
import matplotlib.pyplot as plt
import seaborn as sns

inclinacion_promedio = arboles.groupby('nombre_com')['inclinacio'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=inclinacion_promedio.index, y=inclinacion_promedio.values)
plt.title("Especies con mayor Inclinación promedio")
plt.xlabel("Especies")
plt.ylabel("Inclinación promedio")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
inclinacion_promedio = arboles.groupby('nombre_com')['altura_tot'].mean().sort_values(ascending=True).head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=inclinacion_promedio.index, y=inclinacion_promedio.values)
plt.title("Especies con menor Altura promedio")
plt.xlabel("Especies")
plt.ylabel("Altura promedio")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

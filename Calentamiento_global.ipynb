````python
#### a) Evolución del indicador de cambio climático en la Argetina
Argentina = df_recortada[df_recortada['Country']=='Argentina']
Argentina.index = [""]
Argentina_evolucion = Argentina.drop('Country', axis=1)
Argentina_evolucion.T.plot(title='Evolución del Indicador de Cambio Climático en Argentina')
plt.fill()
plt.legend()
plt.xlabel('Años')
plt.ylabel('Indicador Cambio Climático')
plt.show()
````

````python
#### b)
paises = ['Argentina', 'Chile', 'Uruguay', 'Peru']
Argentina = df_recortada[df_recortada['Country']== paises[0]].drop('Country', axis=1)
Chile = df_recortada[df_recortada['Country']== paises[1]].drop('Country', axis=1)
Uruguay = df_recortada[df_recortada['Country']== paises[2]].drop('Country', axis=1)
Peru = df_recortada[df_recortada['Country']== paises[3]].drop('Country', axis=1)
````

````python
Argentina.index = ["Argentina"]
Chile.index = ["Chile"]
Uruguay.index = ["Uruguay"]
Peru.index = ["Peru"]
Argentina_evolucion = Argentina.iloc[0, 1:]
Chile_evolucion = Chile.iloc[0, 1:]
uruguay_evolucion = Uruguay.iloc[0, 1:]
Peru_evolucion = Peru.iloc[0, 1:]
plt.figure(figsize=(15, 6))
Argentina_evolucion.T.plot(title='Evolución Comparativa del ICC de paises cercanos')
Chile_evolucion.T.plot()
uruguay_evolucion.T.plot()
Peru_evolucion.T.plot()
plt.legend()
plt.xlabel('Año')
plt.ylabel('Indicador de cambio climático')
plt.show()
````

````python
### c) Tendencias entre paises de distinta latitud y longitud

paises_varios = ['Colombia', 'Russian Federation', 'Maldives', 'Uganda', 'Argentina']
Colombia = df_recortada[df_recortada['Country']== paises_varios[0]].drop('Country', axis=1)
Rusia = df_recortada[df_recortada['Country']== paises_varios[1]].drop('Country', axis=1)
Maldivas = df_recortada[df_recortada['Country']== paises_varios[2]].drop('Country', axis=1)
Uganda = df_recortada[df_recortada['Country']== paises_varios[3]].drop('Country', axis=1)
Argentina = df_recortada[df_recortada['Country']== paises_varios[4]].drop('Country', axis=1)
Colombia.index = ["Colombia"]
Rusia.index = ["Rusia"]
Maldivas.index = ["Maldivas"]
Uganda.index = ["Uganda"]
Argentina.index = ["Argentina"]
````

````python
Colombia_evolucion = Colombia.iloc[0, 1:]
Rusia_evolucion = Rusia.iloc[0, 1:]
Maldivas_evolucion = Maldivas.iloc[0, 1:]
Uganda_evolucion = Uganda.iloc[0, 1:]
Argentina_evolucion = Argentina.iloc[0, 1:]

plt.figure(figsize=(15, 6))
Colombia_evolucion.T.plot(title='Evolución Comparativa del ICC')
Rusia_evolucion.T.plot()
Maldivas_evolucion.T.plot()
Uganda_evolucion.T.plot()
Argentina_evolucion.T.plot()
plt.legend()
plt.xlabel('Año')
plt.ylabel('Indicador de cambio climático')
plt.show()
````

En el gráfico "Evolución Comparativa del ICC" podemos observar la comparacion entre paises de distintos continentes y distinta longitud y latitud.

````python
df = df.fillna(0)
df_mayor=df[['Country', 'F2022' ]]
df_mayor= df_mayor.sort_values(by="F2022", ascending=True)
print(f"{df_mayor.Country.iloc[0]} tiene el menor ICC con {df_mayor.F2022.iloc[0]}")
print(f"{df_mayor.Country.iloc[-1]} tiene el mayor ICC con {df_mayor.F2022.iloc[-1]}")
````

````python
df_mayor_dif = df[['Country', 'F1961','F2022']]
df_mayor_dif = df_mayor_dif.loc[(df_mayor_dif['F1961'] != 0) & (df_mayor_dif['F2022'] != 0)]
df_mayor_dif["Diferencia"] = df_mayor_dif["F2022"] - df_mayor_dif["F1961"]
mayor_dif =df_mayor_dif.sort_values(by="Diferencia", ascending=False)
print(f"{mayor_dif.Country.iloc[0]} es el país que mas aumentó el ICC, de {mayor_dif.F1961.iloc[0]} a {mayor_dif.F2022.iloc[0]}")
print(f"{mayor_dif.Country.iloc[-1]} es el país que mas redujo el ICC, de {mayor_dif.F1961.iloc[-1]} a {mayor_dif.F2022.iloc[-1]}")
````

````python
plt.figure(figsize=(18, 6))

Argentina_ICC = df_recortada_Argentina.iloc[0, 1:]

plt.plot(Argentina_ICC)
plt.title('Evolucion del ICC en Argentina')
plt.xlabel('Años')
plt.ylabel('ICC')
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.tight_layout()
plt.margins(x=0)
plt.show()
````

````python
### b)
Argentina_ICC_max_year = Argentina_ICC.idxmax()[1:]
Argentina_ICC_max_value = Argentina_ICC.max()
print(f"El ICC más alto en Argentina fue de: {Argentina_ICC_max_value} en el año {Argentina_ICC_max_year}")
````

````python
Argentina_ICC_min_year = Argentina_ICC.idxmin()[1:]
Argentina_ICC_min_value = Argentina_ICC.min()
print(f"El ICC más bajo en Argentina fue de: {Argentina_ICC_min_value} en el año {Argentina_ICC_min_year}")
````

````python
### c)
Argentina_ICC_Repetidos =(Argentina_ICC.duplicated().sum())
if Argentina_ICC_Repetidos == 0:
  print("No hay valores repetidos de ICC en Argentina")
else:
  print(f"Hay {Argentina_ICC_Repetidos} valores repetidos en Argentina")
````

````python
### d)
Argentina_ICC_promedio = Argentina_ICC.mean()
print(f"El ICC promedio en Argentina fue de: {Argentina_ICC_promedio:.3f}")
````

El ICC promedio en Argentina es la casi la mitad del último registrado (año 2022), pero no refleja la evolución del mismo, es decir hubo registros muy bajos y muy altos.

Escribite un resumen de cómo se usaba el **.loc** y el **.iloc**  asi como el **.max()** , **.min()** y **.mean()**

.loc[]   

Se utiliza para seleccionar datos de un Dataframe. El filtrado se realiza generalmente indicando las etiquetas de las filas y/o columnas. También se puede aplicar una o varias condiciones logicas sobre una o mas columnas.


.iloc[]

Igual que el anterior pero el filtrado se realiza indicando la posición (un entero) de las filas y/o columnas.

.max()

Devuelve el valor más grande.

.min()

Devuelve el valor más chico.

.mean()

Devuelve el promedio.

e) Comparar el indicador de la Argentina de 2022 respecto de el promedio de lo que ocurrió en el mundo. ¿Cuál es mayor?

````python
### e)
Mundo_ICC_promedio = df_recortada[df_recortada['Country']=='World'].iloc[-1,1:].mean()  ### iloc[-1,1:] -1 selecciona la ultima fila y 1: desde la columna 1 hasta el final
Argentina_ICC_2022 = df_recortada_Argentina['F2022'].values[0]  ### en vez de .values[0] puede ser .iloc[0]

print(f"El ICC promedio en el mundo fue de: {Mundo_ICC_promedio:.3f}")
print(f"El ICC en Argentina del 2022 fue de: {Argentina_ICC_2022:.3f}")

if Argentina_ICC_2022 > Mundo_ICC_promedio:
  print("El ICC en Argentina del 2022 es mayor que el promedio mundial.")
else:
  print("El ICC en Argentina del 2022 es menor o igual que el promedio mundial.")
````

```python
import pandas as pd
df = pd.read_csv("/content/Encuesta_Respuestas.txt", sep='\t',encoding='latin-1')
df.head()
```

```python
redes_sociales = df.copy()
redes_sociales['Redes'] = redes_sociales['Redes'].str.split(",")
```

```python

print(redes_sociales.head())
```

```python
filtro_preferencia1 = redes_sociales[(redes_sociales["Edad"]<=18)].copy()
print(filtro_preferencia1)
print(filtro_preferencia1.isin(["Instagram"]))
instagram_filtro1 = filtro_preferencia1.Redes.loc[:] if "Instagram" in filtro_preferencia1.Redes else None
print(instagram_filtro1)
```

```python
instagram_edad1 = filtro_preferencia1[filtro_preferencia1['Redes'].apply(lambda x: 'Instagram' in x)]
print(instagram_edad1.isin(["Instagram"]))
```

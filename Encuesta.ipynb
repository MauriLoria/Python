```python
genero_mujer = redes_sociales[(redes_sociales["Genero"] == "Mujer")].copy()
genero_varon = redes_sociales[(redes_sociales["Genero"] == "Varon")].copy()
genero = genero_mujer.merge(genero_varon, how='outer', on='Genero').value_counts('Genero')
print(genero)
#print(genero_mujer.head(), genero_varon.head())
```

```python
import seaborn as sns
plt.figure(figsize=(12,5))

plt.subplot(121)

redes_sociales_redes = redes_sociales.explode('Redes')  ### La funcion explode lo hace iterable

redes_sociales_redes['Redes'] = redes_sociales_redes['Redes'].str.strip() ### strip() elimina espacio/s en blanco.

g = sns.countplot(data=redes_sociales_redes, x='Redes', hue='Genero')
for container in g.containers:
    g.bar_label(container)    ### Indica la cantidad en la cima de la barra
plt.title("Redes Sociales por Género")
plt.xlabel("Redes Sociales")
plt.ylabel("Cantidad de Usuarios")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.subplot(122)
plt.pie(genero, labels=genero.index, autopct='%1.1f%%', startangle=90) ### autopct muestra los porcentajes en el grafico y stantangle el angulo de inicio
plt.title('Géneros')
plt.axis('equal')

plt.show()
```

```python
orden_uso = ['Por la mañana', 'Por la tarde', 'Por la noche', 'Otro']
plt.figure(figsize=(12,5))

plt.subplot(121)
g = sns.countplot(data=genero_mujer, x='Horario', hue='Uso', order=orden_uso)
for container in g.containers:
    g.bar_label(container)
plt.title("Redes Sociales Usadas por Mujeres")
plt.xlabel("Redes Sociales")
plt.ylabel("Cantidad de Usuarios")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.subplot(122)
g = sns.countplot(data=genero_varon, x='Horario', hue='Uso', order=orden_uso)
for container in g.containers:
    g.bar_label(container)
plt.title("Redes Sociales Usadas por Hombres")
plt.xlabel("Redes Sociales")
plt.ylabel("Cantidad de Usuarios")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.show()
```

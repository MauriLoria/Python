```python
#cambiar la duracion a numérico formato tiempo
#Hacer plot
df=df.dropna(subset='Duración')
from datetime import datetime

duracion=[]
for i in df['Duración']:
  a=datetime.strptime(i, "%M:%S").time()
  duracion.append(a)

df['Duración']=duracion
df
```

```python
#Graficar
ax1 = df.plot(x='Duración',y='Popularidad')
```

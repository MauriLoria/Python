```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("/content/Encuesta_Respuestas.txt", sep='\t',encoding='latin-1')
df.head()
```

```python
redes_sociales = df.copy()
redes_sociales['Redes'] = redes_sociales['Redes'].str.split(",")
```

```python
edad1 = redes_sociales[(redes_sociales["Edad"]<=18)].copy()
edad2 = redes_sociales[(redes_sociales["Edad"]>18) & (redes_sociales['Edad']<=30)].copy()
edad3 = redes_sociales[(redes_sociales["Edad"]>30) & (redes_sociales['Edad']<=40)].copy()
edad4 = redes_sociales[(redes_sociales["Edad"]>40)].copy()
instagram_edad1 = edad1[edad1['Redes'].apply(lambda x: any("Instagram" in item.strip() for item in x))]   ### strip() elimina espacio/s en blanco.
whatsapp_edad1 = edad1[edad1['Redes'].apply(lambda x: any("Whatsapp" in item.strip() for item in x))]   ### strip() elimina espacio/s en blanco.
facebook_edad1 = edad1[edad1['Redes'].apply(lambda x: any("Facebook" in item.strip() for item in x))]   ### strip() elimina espacio/s en blanco.
tiktok_edad1 = edad1[edad1['Redes'].apply(lambda x: any("Tik Tok" in item.strip() for item in x))]   ### strip() elimina espacio/s en blanco.
otro_edad1 = edad1[edad1['Redes'].apply(lambda x: any("Otro" in item.strip() for item in x))]   ### strip() elimina espacio/s en blanco.
instagram_edad2 = edad2[edad2['Redes'].apply(lambda x: any("Instagram" in item.strip() for item in x))]   ### strip() elimina espacio/s en blanco.
whatsapp_edad2 = edad2[edad2['Redes'].apply(lambda x: any("Whatsapp" in item.strip() for item in x))]   ### strip() elimina espacio/s en blanco.
facebook_edad2 = edad2[edad2['Redes'].apply(lambda x: any("Facebook" in item.strip() for item in x))]   ### strip() elimina espacio/s en blanco.
tiktok_edad2 = edad2[edad2['Redes'].apply(lambda x: any("Tik Tok" in item.strip() for item in x))]   ### strip() elimina espacio/s en blanco.
otro_edad2 = edad2[edad2['Redes'].apply(lambda x: any("Otro" in item.strip() for item in x))]   ### strip() elimina espacio/s en blanco.
instagram_edad3 = edad3[edad3['Redes'].apply(lambda x: any("Instagram" in item.strip() for item in x))]   ### strip() elimina espacio/s en blanco.
whatsapp_edad3 = edad3[edad3['Redes'].apply(lambda x: any("Whatsapp" in item.strip() for item in x))]   ### strip() elimina espacio/s en blanco.
facebook_edad3 = edad3[edad3['Redes'].apply(lambda x: any("Facebook" in item.strip() for item in x))]   ### strip() elimina espacio/s en blanco.
tiktok_edad3 = edad3[edad3['Redes'].apply(lambda x: any("Tik Tok" in item.strip() for item in x))]   ### strip() elimina espacio/s en blanco.
otro_edad3 = edad3[edad3['Redes'].apply(lambda x: any("Otro" in item.strip() for item in x))]   ### strip() elimina espacio/s en blanco.
instagram_edad4 = edad4[edad4['Redes'].apply(lambda x: any("Instagram" in item.strip() for item in x))]   ### strip() elimina espacio/s en blanco.
whatsapp_edad4 = edad4[edad4['Redes'].apply(lambda x: any("Whatsapp" in item.strip() for item in x))]   ### strip() elimina espacio/s en blanco.
facebook_edad4 = edad4[edad4['Redes'].apply(lambda x: any("Facebook" in item.strip() for item in x))]   ### strip() elimina espacio/s en blanco.
tiktok_edad4 = edad4[edad4['Redes'].apply(lambda x: any("Tik Tok" in item.strip() for item in x))]   ### strip() elimina espacio/s en blanco.
otro_edad4 = edad4[edad4['Redes'].apply(lambda x: any("Otro" in item.strip() for item in x))]   ### strip() elimina espacio/s en blanco.
#print("La Cantidad de mayores de 18 a単os que usan Instagram son: ", instagram_edad3.Redes.count())
#print("La Cantidad de mayores de 18 a単os que usan Whatsapp son: ", whatsapp_edad3.Redes.count())
#print("La Cantidad de mayores de 18 a単os que usan Facebook son: ", facebook_edad3.Redes.count())
#print("La Cantidad de mayores de 18 a単os que usan Tik Tok son: ", tiktok_edad3.Redes.count())
#print("La Cantidad de mayores de 18 a単os que usan otras Redes Sociales son: ", otro_edad3.Redes.count())
redes = ["Instagram", "Whatsapp", "Facebook", "Tik Tok", "Otra"]
cantidad_edad1 = [instagram_edad1.Redes.count(), whatsapp_edad1.Redes.count(), facebook_edad1.Redes.count(), tiktok_edad1.Redes.count(), otro_edad1.Redes.count()]
cantidad_edad2 = [instagram_edad2.Redes.count(), whatsapp_edad2.Redes.count(), facebook_edad2.Redes.count(), tiktok_edad2.Redes.count(), otro_edad2.Redes.count()]
cantidad_edad3 = [instagram_edad3.Redes.count(), whatsapp_edad3.Redes.count(), facebook_edad3.Redes.count(), tiktok_edad3.Redes.count(), otro_edad3.Redes.count()]
cantidad_edad4 = [instagram_edad4.Redes.count(), whatsapp_edad4.Redes.count(), facebook_edad4.Redes.count(), tiktok_edad4.Redes.count(), otro_edad4.Redes.count()]
```

```python
plt.figure(figsize=(12,5))

plt.subplot(121)
plt.title("Menores de 18 a単os y las Redes Sociales")
barras=plt.bar(redes, cantidad_edad1)
plt.bar_label(barras)
plt.xlabel("Redes Sociales (Figura N属 1)")
plt.ylabel("Cantidad")

plt.subplot(122)
plt.title('Mayores de 18 y menores de 30 a単os y las Redes Sociales')
barras=plt.bar(redes, cantidad_edad2)
plt.bar_label(barras)
plt.xlabel("Redes Sociales (Figura N属 2)")
plt.ylabel("Cantidad")

plt.show()

plt.figure(figsize=(12,5))

plt.subplot(121)
plt.title('Mayores de 30 y menores de 40 a単os y las Redes Sociales')
barras=plt.bar(redes, cantidad_edad3)
plt.bar_label(barras)
plt.xlabel("Redes Sociales (Figura N属 3)")
plt.ylabel("Cantidad")

plt.subplot(122)
plt.title('Mayores de 40 a単os y las Redes Sociales')
barras=plt.bar(redes, cantidad_edad4)
plt.bar_label(barras)
plt.xlabel("Redes Sociales (Figura N属 4)")
plt.ylabel("Cantidad")

plt.show()
```
